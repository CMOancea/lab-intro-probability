**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
!pip install scipy
```

```python
from scipy.stats import binom

# Info
n = 460          # Tickets solded
p = 0.97         # probability of a persona showing up
seats = 450   # Max capacity 

# We wont to know the probability of 450 or fewer people coming (P <= 450)
probability = binom.cdf(seats, n, p)

print(f"the probability of having seats for everyone is: {probability:.4f}")
```

**Challenge 2**

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python
from scipy.stats import poisson

# Parameters
# Lambda (average rate of complaints per hour)
lambda_rate = 4
# k (the exact number of events we are interested in)
k = 6

# Calculate the probability of exactly k events using the PMF
probability = poisson.pmf(k, lambda_rate)

print(f"The probability of exactly 6 calls is: {probability:.4f}")
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
from scipy.stats import poisson

# Parameters
# Lambda (average rate of visits per hour)
lambda_rate = 500
# The maximum capacity the server can handle without being overwhelmed
capacity_limit = 550

# The probability of being overwhelmed is P(X > 550).
# We calculate the complement: P(X <= 550) using the Cumulative Distribution Function (CDF).
prob_not_overwhelmed = poisson.cdf(capacity_limit, lambda_rate)

# The probability of being overwhelmed is 1 minus the complement
prob_overwhelmed = 1 - prob_not_overwhelmed

print(f"The probability of the website server being overwhelmed is: {prob_overwhelmed:.4f}")
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
from scipy.stats import poisson

# Parameters
lambda_rate = 500
capacity_limit = 550
hours_in_day = 24

# 1. Calculate P(overwhelmed) for a single hour (P_fail)
# P(fail in 1 hour) = 1 - P(X <= 550)
P_not_overwhelmed_in_1hr = poisson.cdf(capacity_limit, lambda_rate)
P_overwhelmed_in_1hr = 1 - P_not_overwhelmed_in_1hr

# 2. Calculate P(no overwhelming) for 24 consecutive hours (P_success^24)
P_no_overwhelming_in_24hrs = P_not_overwhelmed_in_1hr ** hours_in_day

# 3. Calculate P(at least one overwhelming event)
P_at_least_one_overwhelming_in_24hrs = 1 - P_no_overwhelming_in_24hrs

print(f"P(overwhelmed in 1 hour): {P_overwhelmed_in_1hr:.4f}")
print(f"P(at least one overwhelming in 24 hours): {P_at_least_one_overwhelming_in_24hrs:.4f}")
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
from scipy.stats import expon

# Parameters
# Average time between arrivals is 10 minutes (mean).
# In the Exponential distribution, the 'scale' parameter is the mean (1/lambda).
mean_time_between_arrivals = 10 
scale_parameter = mean_time_between_arrivals 

# Time (t) we are interested in
t = 5 

# We want P(T <= 5), which is the Cumulative Distribution Function (CDF) at t=5.
# The expon.cdf uses the rate implicitly via the scale parameter.
probability = expon.cdf(t, scale=scale_parameter)

print(f"The probability of the next customer arriving within the next 5 minutes is: {probability:.4f}")
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
from scipy.stats import expon

# Parameters
# Average time between arrivals (mean), which is the 'scale' parameter for scipy
mean_time_between_arrivals = 10 
scale_parameter = mean_time_between_arrivals 

# Minimum time required without a customer to take a break
required_break_time = 15 

# We want P(T >= 15), which is the Survival Function (SF) at t=15.
probability = expon.sf(required_break_time, scale=scale_parameter)

print(f"The probability of the next customer arriving after 15 minutes (allowing a break) is: {probability:.4f}")
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
from scipy.stats import norm

# Parameters for the Normal Distribution
mean = 150
std_dev = 10

# The range of interest
lower_bound = 140
upper_bound = 160

# We calculate P(X <= 160) using the Cumulative Distribution Function (CDF)
P_up_to_160 = norm.cdf(upper_bound, loc=mean, scale=std_dev)

# We calculate P(X <= 140) using the CDF
P_up_to_140 = norm.cdf(lower_bound, loc=mean, scale=std_dev)

# The probability P(140 < X < 160) is the difference between the two CDFs
probability = P_up_to_160 - P_up_to_140

print(f"P(X <= 160): {P_up_to_160:.4f}")
print(f"P(X <= 140): {P_up_to_140:.4f}")
print(f"The probability that the weight is between 140 and 160 grams is: {probability:.4f}")
```

**Challenge 6**

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
from scipy.stats import expon

# Parameters
# Mean lifetime (mu), which is the 'scale' parameter for scipy
mean_lifetime = 50 
scale_parameter = mean_lifetime 

# Time (t) we are interested in
t = 30 

# We want P(T <= 30), which is the Cumulative Distribution Function (CDF) at t=30.
probability = expon.cdf(t, scale=scale_parameter)

print(f"The probability that the component fails within the first 30 hours is: {probability:.4f}")
```
